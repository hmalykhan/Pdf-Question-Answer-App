```python
from llama_index.legacy import VectorStoreIndex,SimpleDirectoryReader
documents=SimpleDirectoryReader("data").load_data()
```

```python
index=VectorStoreIndex.from_documents(documents,show_progress=True)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
from llama_index.retrievers import VectorIndexRetriever
from llama_index.query_engine import RetrieveQueryEngine
from llama_index.indices.postprocessor import SimilarityPostprocessor
retriever=VectorIndexRetriever(index=index,Similarity_top_k=4)
postprocessor=SimilarityPostprocessor(similarity_cutoff=0.80)
query_engine=RetrieveQueryEngine(retriever=retriever,node_postprocessors=postprocessor)
```

```python
response=query_engine.query("what is the transformer")
```

```python
from llama_index.response.pprints_utils import pprint_response
print(response)
```
